## 품질데이터 가져오기

```python
def get_quality_data_list_for_predict():
    conn = pymysql.connect(host='127.0.0.1', user='root', password='carry789', db='projectdata', charset='utf8')

    sql = '''
        select product_key, product_size_l, product_size_w, product_size_h
        from product_quality
    '''

    cursor = conn.cursor()
    cursor.execute(sql)
    row = cursor.fetchall()

    data_list = []

    for obj in row:
        data_dic = {
            'product_key': obj[0],
            'product_size_l': obj[1],
            'product_size_w': obj[2],
            'product_size_h': obj[3]
        }
        data_list.append(data_dic)

    conn.close()

    return data_list
```

```python
quality_data = get_quality_data_list_for_predict()
```

```python
quality_df = pd.DataFrame(quality_data)
```

## 설비 데이터 가져오기

```python
def get_machine_data_list_for_predict():
    conn = pymysql.connect(host='127.0.0.1', user='root', password='carry789', db='projectdata', charset='utf8')

    sql = '''
        select machine_data, machine_data_code
        from machine
    '''

    cursor = conn.cursor()
    cursor.execute(sql)
    row = cursor.fetchall()

    data_list = []

    for obj in row:
        data_dic = {
            'machine_data': obj[0],
            'machine_data_code': obj[1]
        }
        data_list.append(data_dic)

    conn.close()

    return data_list
```

```python
machine_data = get_machine_data_list_for_predict()
```

```python
machine_df = pd.DataFrame(machine_data)
```
